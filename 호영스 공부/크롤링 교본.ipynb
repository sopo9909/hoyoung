import 모듈

→ 해당 모듈 전체를 가져온다.

    사용하려면 항상 '모듈명.메소드' 와 같이 모듈명을 앞에 붙여주어야 한다.



from 모듈 import 메소드 / 변수

→ 해당 모듈 내에 있는 특정 메소드나 모듈 내 정의된 변수를 가져온다.

    가져온 메소드나 변수를 앞에 모듈명을 붙이지 않고 그대로 사용할 수 있다.

    다만, 이름이 같은 변수나 메소드가 존재할 경우 대체된다.

    

    from 모듈 import * 이라고 하면 import 모듈과 동일하다. (사용 시 모듈명 붙이는 것 빼고)

### 실전 2번

```python
import re
import requests
from bs4 import BeautifulSoup

URL='https://search.musinsa.com/category/001'
response=requests.get(URL)
soup=BeautifulSoup(response.text,'html.parser')
result_brand=soup.select('div.section_container_list p.item_title>a')
result_name=soup.select('.list_info>a[name="goods_link"]')
result_price=soup.select('div[id="goods_list"] p.price')

brand=[]
for i in result_brand:
    brand.append(i.text)
#print(len(brand))

name=[]
for i in result_name:
    name.append(re.sub(r'[\s]','',i.text))
#print(len(name))

price=[]
for i in result_price:   #i는 태그형태
    if len(i)>=2:
        price_2=re.sub(r'["\s"]','',i.text)
        price.append(price_2[1])
    else:
        price.append(re.sub(r'["\s"]','',i.text))
        
for i in range(len(brand)):
    print('{}:{}'.format('브랜드',brand[i]))
    print('{}:{}'.format('제품명',name[i]))
    print('{}:{}'.format('가격',price[i]),'\n')
```

### 실전 2번

```python
import re
import requests
from bs4 import BeautifulSoup
url="https://search.musinsa.com/category/001"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
list_items = soup.select('div.list-box ul#searchList li.li_box div.article_info')
#brands = soup.select('div.list-box ul#searchList li.li_box p.item_title')
keys = ["브랜드", "제품명", "가격"]
product_list = []
for item in list_items:
    brands = item.find('p', attrs={"class":"item_title"})
    products = item.find('a', attrs={"name":"goods_link"})
    prices = item.find('p', attrs={"class":"price"})
    
    for brand in brands:
        product_brand = brand.text
    
    for product in products:
        #product.replace('<strong class="txt_reserve">12/31 배송 </strong>', " ")
        try:
            product_name = product.split()
        except TypeError:
            pass
        a = " "
        product_name = a.join(product_name)
    
    for price in prices:
        try:
            product_price = price.split()
        except:
            pass
        product_price = a.join(product_price)
    product = [product_brand,product_name,product_price]
    product_list.append(dict(zip(keys, product)))
product_list
```
