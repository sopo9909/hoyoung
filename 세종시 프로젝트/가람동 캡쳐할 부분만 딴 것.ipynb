```python
data = pd.read_csv('3.세종시_아파트(매매)_실거래가.csv')
```

```python
#동 구분하는 함수
def sort_dong2(df):
    a=df.split(" ")
    return a[1]
data['읍면동']=data['시군구'].apply(sort_dong2)
```

```python
def year(df):
    return df[:-2]
data['계약년월']=data['계약년월'].astype(str)
data['계약년']=data['계약년월'].apply(year)
```

```python
def riri(df):
    return int(float(df.replace(',','')))
data['거래금액(만원)'] =data['거래금액(만원)'].apply(riri)
```
