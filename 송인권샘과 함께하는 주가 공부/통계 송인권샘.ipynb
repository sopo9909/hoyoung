```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ri=[]
for i in range(100):
    r=int(np.random.normal(0,5)*10)
    ri.append(r)
print(ri)
```

```python
plt.hist(ri)
plt.show()
```

```python
ys = np.random.normal(0,1,1000)
plt.hist(ys)
```

```python
#평균
rmean=np.mean(ri)
#중앙값
rmedian=np.median(ri)
#사분위수
q1=np.percentile(ri,25)
q2=np.percentile(ri,50)
q3=np.percentile(ri,75)
#분산
var=np.var(ri)
std=np.std(ri)

[rmean,rmedian,q1,q2,q3,var,std]
plt.boxplot(ri)
plt.show()
```

```python
fig, ax = plt.subplots()
#notch : 중위값 강조 , vert: 회전 ,sym : 극단치
ax.boxplot([ri,ys],positions=[2,2.5],notch=1,vert=0,sym='ro')
plt.show()
```

```python
a=3
b=5
xs=np.random.normal(0,5,10000)
minx=np.min(xs)
maxx=np.max(xs)
miny=a*minx+b
maxy = a*maxx+b
ys=[]
data=[]
for x in xs:
    e=np.random.normal(0,3)
    y=a*x+b+e
    ys.append(y)
    data.append((x,y))
plt.scatter(xs,ys)
plt.plot([minx,maxx],[miny,maxy],'r+')
plt.show()
```

```python
#data = [(x,y)]
def makeMSE(a,b,data):
    temp=[]
    elist=[]
    for dx,dy in data:
        yh=a*dx+b
        e=(yh-dy)**2
        elist.append(e)
    mse=np.sum(elist)/len(data)
    return mse
        
    #mse=sum(np.mean(yh-yc)**2)/len(y)
```

```python
mse=makeMSE(2,3,data)
mse
```

```python
def nextMSE(a ,b ,data,lr=0.01) :
    bmse=makeMSE(a,b,data)
    amse=makeMSE(a+lr,b,data)
    if(amse < bmse):
        return a+lr
    else:
        return a-lr
```

```python
resa=[]
a=2
b=3
for i in range(1000):
    a=nextMSE(a,b,data)
    resa.append(a)
resa
```
