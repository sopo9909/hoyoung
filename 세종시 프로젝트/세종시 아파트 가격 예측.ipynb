```python
#읍면동 구분하는 함수
def sort_dong(df):
    a=df.split(" ")
    if a[1][-1]=='동':
        return 3
    elif a[1][-1]=='면':
        return 1
    else :
        return 2
def sort_dong3(df):
    if df=='동':
        return 3
    elif df=='면':
        return 1
    else :
        return 2
#동 구분하는 함수
def sort_dong2(df):
    a=df.split(" ")
    return a[1]
#년도 추출
def year(df):
    return int(df[:-2])
def mon(df):
    return int(df[-2:])
def riri(df):
    return int(float(df.replace(',','')))
```

```python
def dongnum(df):
    if df=='보람동':
        return 10
    elif df=='종촌동':
        return 11
    elif df=='고운동':
        return 12 
    elif df=='아름동':
        return 13 
    elif df=='어진동':
        return 14 
    elif df=='연서면':
        return 8 
    elif df=='조치원읍':
        return 9 
    elif df=='한솔동':
        return 15 
    elif df=='도담동':
        return 16
    elif df=='연동면':
        return 1 
    elif df=='장군면':
        return 2 
    elif df=='나성동':
        return 17 
    elif df=='부강면':
        return 3 
    elif df=='소담동':
        return 18 
    elif df=='금남면':
        return 4 
    elif df=='전의면':
        return 5 
    elif df=='전동면':
        return 6 
    elif df=='소정면':
        return 7
    elif df=='새롬동':
        return 19 
    elif df=='다정동':
        return 20 
    elif df=='대평동':
        return 21 
    elif df=='반곡동':
        return 22 
    elif df=='가람동':
        return 23 
    elif df=='해밀동':
        return 24
```

```python
data = pd.read_csv('3.세종시_아파트(매매)_실거래가.csv')
```

```python
data['읍면동구분']=data['시군구'].apply(sort_dong)
data['읍면동']=data['시군구'].apply(sort_dong2)
data['읍면동num']=data['읍면동'].apply(dongnum)
data['계약년월']=data['계약년월'].astype(str)
data['월']=data['계약년월'].apply(mon)
data['년']=data['계약년월'].apply(year)
data['거래금액2(만원)']= data['거래금액(만원)'].apply(riri)
labelencoder_x= LabelEncoder()
y = data["단지명"]
data["단지명2"]=labelencoder_x.fit_transform(y)
```
