```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import requests
import re
import sqlite3
options = webdriver.ChromeOptions()
#options.add_argument('headless')
options.add_argument("start-maximized")
options.add_argument("disable-infobars")
options.add_argument("--disable-extensions")
options.add_argument('lang=ko_KR')
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36')
```

```python
driver = webdriver.Chrome(chrome_options=options,executable_path="chromedriver.exe")
```

```python
url = 'https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000000444&dispCatNo=1000001000500030001&trackingCd=Drawer_Cat100000100050003_Cat'
driver.get(url)
driver.implicitly_wait(3)
driver.find_element_by_css_selector('#reviewInfo').click()
```

```python
for p in range(2,11):
    html = driver.page_source
    htmls = BeautifulSoup(html,'html.parser')
    information_list = htmls.find_all(attrs={'class':'review_cont'})
    for information in information_list:
        r=information.find(attrs={'class':'txt_inner'})
        re=information.find_all(attrs={'class':'txt'})
        rate=information.find(attrs={'class':'point'}).text[-2:]
        print(r.text)
        print(re)
        print(rate)
    driver.find_element_by_css_selector('div.pageing > a[data-page-no="{0}"]'.format(p)).click()
    #txt_inner
```
