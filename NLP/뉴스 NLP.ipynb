```python
def getSub(whr=''):
    dbname='navernews.db'
    with sqlite3.connect(dbname) as conn:
        cursor=conn.cursor()
        sql ='select * from subnews'+whr
        res=cursor.execute(sql).fetchall()
        return res
```
