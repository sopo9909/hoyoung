```python
#안됨
import sqlite3
def insMain(d,db):
    with sqlite3.connect(db) as conn:
        cursor=conn.cursor()
        sql='select * from SSG where review="'+d[2]+'" AND product="'+d[1]+'"'
        res=cursor.execute(sql).fetchall()
        if(len(res)>0):
            print("키가 이미 존재 합니다.:"+d[2])
        else:
            sql='insert into SSG (brand,product,review,rate,Frag,Frag_time) values(?,?,?,?,?,?)'
            res=cursor.execute(sql,d)
            print('.')
```

```python
def rat_Frag(width):
    if width =='은은해요':
        return 5
    elif width =='보통이에요':
        return 3
    elif width =='강해요':
        return 1
def rat_Time(width):
    if width =='오래가요':
        return 5
    elif width =='보통이에요':
        return 3
    elif width =='예상보다 짧아요':
        return 1
```

```python
def rating(width):
    if width =='100%':
        return 5
    elif width =='90%':
        return 4.5
    elif width =='80%':
        return 4
    elif width =='70%':
        return 3.5
    elif width =='60%':
        return 3
    elif width =='50%':
        return 2.5
    elif width =='40%':
        return 2
    elif width =='30%':
        return 1.5
    elif width =='20%':
        return 1
    elif width =='90%':
        return 0.5
    else:
        return 0
```
